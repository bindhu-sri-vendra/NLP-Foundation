```python
#simple function that takes any text divides into words and provides length of those words
def word_count(text):
    word = text.split()
    print('word count:',len(word))

word_count(text)
```

```python
# word count using regex
import re

def count_words(text):
    words = re.findall(r'\b\w+\b', text)
    print('word count:', len(words))

count_words(text)
```

```python
#full proof method to count the words

from nltk import word_tokenize 

def count_words_nltk(text):
    words = word_tokenize(text)
    return len(words)

text = "Let's explore NLP with python ."

print(word_tokenize(text))
count_words_nltk(text)
```

```python
#spacy implementation

import spacy

nlp = spacy.load('en_core_web_sm')

def count_words_spacy(text):
    doc = nlp(text)
    words = [token.text for token in doc if not token.is_punct]
    return len(words)

text = "Let's explore NLP with python ."
count_words_spacy(text)

doc = nlp(text)
print([token.text for token in doc])
```

```python
#to check word frequency 
from collections import Counter

def word_frequency(text):
    words = text.lower().split()
    return Counter(words)

text = "NLP is fun, NLP is powerful, and nlp is amazing in the future."
print(word_frequency(text))
```

```python
def word_frequency(text):
    doc = nlp(text)
    words = [token.text.lower() for token in doc if not token.is_punct]
    return Counter(words)

text = "NLP is fun, NLP is powerful, and nlp is amazing in the future."
print(word_frequency(text))
```

```python
#NLP - Task 1 - short vs long sentences
from nltk.tokenize import sent_tokenize, word_tokenize

def categorize_sentences(text):
    sentences = sent_tokenize(text)
    categorized = {'short':[],'long':[]}

    for sentence in sentences:
        doc = nlp(sentence)
        word_count = len([token for token in doc if not token.is_punct])
        if word_count <5:
            categorized['short'].append(sentence)
        else:
            categorized['long'].append(sentence)
    return categorized   
```

```python
text = 'Hello, NLP is fun. I like it very much. I hope everyone is following this lecture.'

categories = categorize_sentences(text)
print('short sentences', categories['short'])
print('long sentences', categories['long'])
```

```python
#mini assignment - check if the para has how many question sentences and non question ones
def ques_categorizer(text):
    sentences = sent_tokenize(text) # converts para to individual sentences
    categories = {'questions':[],'not_questions':[]}#we are defining categories dictionary

    #convert each sentence into token 
    for sentence in sentences:
        if '?' in sentence:
            categories['questions'].append(sentence)
        else:
            categories['not_questions'].append(sentence)
    return categories
```

```python
text = 'Hello ?. NLP is fun. I like it very much. I hope everyone is following this lecture ?'

categories = ques_categorizer(text)
print('sentences with questions', categories['questions'])
print('sentences without questions', categories['not_questions'])
```

```python
#rule based sentiment analysis

import time

positive_words = {'happy'}
negative_words = {'sad'}
```

```python
def sentiment_analysis(text):
    start_time = time.time()
    words = text.lower().split()

    pos_count = sum(1 for word in words if word in positive_words)
    neg_count = sum(1 for word in words if word in negative_words)

    sentiment = 'Positive Sentiment' if pos_count > neg_count else 'Negative Sentiment' if neg_count > pos_count else 'Neutral'
    end_time = time.time()
    execution_time = end_time - start_time

    return sentiment, execution_time
```

```python
text = 'I am very happy'

sentiment, execution_time = sentiment_analysis(text)
print(sentiment, execution_time)
```
