```python
from collections import Counter
```

```python
#word count
text = 'I wanna study at kaist because kaist is awesome'

#divide the words from the text
words = text.lower().split()

#count the words 
print(Counter(words))
```

```python
freq = Counter(words)
print(freq.most_common(3)) # the 3 most common words from the text 
```
