```python
import spacy
```

```python
#load the english nlp model
nlp = spacy.load("en_core_web_sm")
```

```python
pos_sentence = 'NLP is amazing'
neu_sentence = 'Google was invented by Larry and Sergy'
```

```python
#POS - parts of speech tagging Annotation
pos_doc = nlp(pos_sentence)
pos_tags = [(token.text, token.pos_)for token in pos_doc]

print(pos_tags)
```

```python
#NER - named entitiy recognition Annotation
ner_doc = nlp(neu_sentence)
ner_tags = [(token.text, token.label_) for token in ner_doc.ents]

print(ner_tags)
```

```python
#example 2 for NER 
text = 'Bindhu is studying at kaist and her fee is 20 million USD in 2027'

exdoc = nlp(text)
tags = [(ent.text, ent.label_) for ent in exdoc.ents]
print(tags)
```
