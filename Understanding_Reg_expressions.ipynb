```python
#we need to import the library called regular expression
import re
```

```python
# to search for a specific pattern in a text 
text = 'I love python programming'
match = re.search(r'python', text)
print(match.group())
```

```python
text = 'hello world'
match = re.search(r'hello.w',text)
print(match.group())
```

```python
#to know the starting of the string 
text = 'hello world'
match = re.search(r'^hello',text)
print(match.group())
```

```python
#to check the ending of the string we use $
match = re.search(r'world$', text)
print(match.group())
```

```python
match = re.search(r'lo*',text)
print(match.group())
```

```python
a = 'helloooo world'
b = 'hell world'
print((re.search(r'llo*',a)).group())
```

```python
print((re.search(r'o*',b)).group()) 
```

```python
# we use + to match 1 or more characters 
match = re.search(r'lo+', text)
print(match.group())
```

```python
text = 'in mye name, XYX, ABC'
match = re.findall(r'[aeiou]', text)
print(match)
```

```python
#if we want to find the characters that are not in []
print(re.findall(r'[^aeiou]', text))
```

```python
text = '1234 and 567'
print(re.findall(r'\d',text)) # it prints all the digits in our text
```

```python
print(re.findall(r'\d{2}',text))
```

```python
print(re.findall(r'\d{3}',text))
```

```python
print(re.findall(r'\d{2,3}',text))
```

```python
print(re.findall(r'\d*',text)) #we will have everything
```

```python
print(re.findall(r'\d+',text)) #only the numbers string
```

```python
text = ' i like sage color and black colour'
#we want to find color and colour too but the spellings are different so we can use ?
print(re.findall(r'colou?r', text)) # the letter before ? becomes optional
```

```python
#grouping and capturing 

text = 'call me at 010-929-2949'
match = re.search(r'(\d{3})-(\d{3})-(\d{2})',text)
print(match.group())
```

```python
#when we want a pattern followed by something
text = 'abc123'
match = re.search(r'abc(?=123)',text)
print(match.group())
```

```python
#when we want a pattern that is not followed by something
match = re.search(r'abc(?!23)',text)
print(match.group())
```

```python
#to match a particular word we use \b 
text = 'i love programming'
match = re.search(r'\bprogramming\b',text)
print(match.group())
```

```python
text = 'my phone number is 1234'
print(re.sub(r'\d+', '4321', text))
```

```python
print(re.sub(r'phone', 'game', text))
```

```python
text = 'hello. how are you?'
match = re.search(r'hello\.',text)
print(match.group())
```

```python
text = 'hello#. how are you?'
match = re.search(r'hello\#.',text)
print(match.group())
```

```python
#to match the digits and non digits 
text = 'there are 3 apples'
```

```python
print(re.findall(r'\d',text))
```

```python
print(re.findall(r'\D',text)) #for non digits
```

```python
print(re.findall(r'\D+',text))
```

```python
print(re.findall(r'\D*',text))
```

```python
#to get all the whitespaces
print(re.findall(r'\s',text))
```

```python
#to get everything that is not a white space 
print(re.findall(r'\S',text))
```

```python
print(re.findall(r'\S+',text))
```

```python
#matching exact string lengths
text = 'abc def dgk'
print(re.findall(r'\b[a-zA-Z]{3}\b',text))
```

```python
print(re.findall(r'\b[a-zA-Z]{2}\b',text)) #wont print anything cause there are no words in our text that are of length 2
```

```python
#to remove all white spaces
text = ' hello  world '
print(re.sub(r'\s','',text))
```

```python
#matching dates 
text = 'event is on 12/12/2025'
match = re.search(r'\d{2}/\d{2}/\d{4}',text)
print(match.group())
```
