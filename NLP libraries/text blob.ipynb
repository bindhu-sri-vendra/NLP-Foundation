```python
#mini task >> take input in any lang and convert it to english and identify the sentiment 

texts = [
    "This product is amazing",
    "이 제품은 별로에요",
    "Este producto es terrible",
    "C'est vraiment bon",
    "I hate this service"
]
```

```python
def translate(texts):
    for text in texts:
        print('original text',text)
        translated = GoogleTranslator(source ='auto', target='en').translate(text)
        print('translated text', translated)
        blob = TextBlob(translated)
        polarity = blob.sentiment.polarity
        print('polarity is',polarity)
        if(polarity <=-1 and polarity>=1):
            print('Neutral Statement') #polarity is rarely 0 so dont use strict condition like polarity == 0
        elif(polarity <0):
            print('Negative Statement')
        else:
            print('Positive Statement')
```

```python
translate(texts)
```

```python
#POS tagging
blob = TextBlob('Barack Obama was born in Hawaii')
for word,tag in blob.tags:
    print(word,tag)
```

```python
#text summarization
from collections import Counter

text = """ Python is a great programming language.
It is widely used for data analytics, machine learning , web development, and automation."""

blob = TextBlob(text)
nouns = [word.lower() for word,tag in blob.tags if tag.startswith('NN')]

common_nouns = Counter(nouns).most_common(5)
summary = ' '.join(word for word,count in common_nouns)
print(summary)
```
