```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
text = 'Apple is looking at buying a startup in the UK for $1 billion.'

doc = nlp(text)

for token in doc:
    print(token.text)
```

```python
for token in doc:
    #print(token.is_stop)
    #print(token.pos_)
    print(token.lemma_)
```

```python
#NER .. named entity recognition
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
#dependency parsing (sentence structure)
for token in doc:
    print(token.text, token.dep_, token.head.text)
```

```python
#noun chunks
for chunk in doc.noun_chunks:
    print(chunk.text)
```

```python
#fast text cleaning function
def clean_text(text):
    doc = nlp(text)
    tokens = []
    for token in doc:
        if not token.is_stop and not token.is_punct:
            tokens.append(token.lemma_.lower()) #lemma_ >> root form 
    return " ".join(tokens)


print(clean_text("Apple is looking at buying startups!"))
```

```python
#convert text to features 
#spacy doesn't do TF-IDF by itself so we combine with sklearnk

from sklearn.feature_extraction.text import TfidfVectorizer

texts = ['I love NLP',
        'NLP is amazing',
        'I hate bugs']
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
print(X.toarray())
```
