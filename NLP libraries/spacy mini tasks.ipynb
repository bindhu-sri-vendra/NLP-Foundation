```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
#Task 1 >> Clean a sentence - remove stopwords , punctuation, convert to lemma and lower case
text = 'I am running towards the big buildings!'

def clean_text(text):
    doc = nlp(text)
    tokens=[]

    for token in doc:
        if not token.is_stop and not token.is_punct:
            tokens.append(token.lemma_.lower())
    return " ".join(tokens)
```

```python
clean_text(text)
```

```python
def cleantxt(text):
    doc = nlp(text)
    tokens=[]

    for token in doc:
        if token.pos_ in ['NOUN', 'PROPN', 'VERB', 'ADJ']: #we use this to not remove words that has actual meaning as stop words
            tokens.append(token.lemma_.lower())
    return " ".join(tokens)

print(cleantxt(text))
```

```python
#task2 - count useful and useless words in a sentence 
total = len (text.split())
useful_words = len(clean_text(text).split())
print('useful words',useful_words, 'useless words', total-useful_words)
```

```python
#task 3 >> print nouns only 
doc =nlp(text)
for token in doc:
    if token.pos_ in ['NOUN', 'PROPN']:
        print(token.text)
```

```python
#task 4 >> extract all noun phrases
#chunk is a group of words acting as a noun
#root is main word of the phrase

text = 'I ordered a large pizza and a cold drink'
doc = nlp(text)
for chunk in doc.noun_chunks:
    print('phrase', chunk.text)
    print('root', chunk.root.text)
```

```python
#task 5 >> NER >> extract person, org, location, date
text = 'Elon Musk founded SpaceX in the USA in 2002'
doc = nlp(text)

for ent in doc.ents:
    print(ent.label_,ent.text)
```

```python
#task 6 >> find subject and verb >> depending parsing
for token in doc:
    if token.dep_ == 'nsubj':
        print('Subject', token.text)
    elif token.dep_ =='ROOT':
        print('Verb', token.text)
```

```python
#task 8 >> key word extraction
text = "Elon Musk announced a new AI product at the Tesla event in California"
doc = nlp(text)

#task is to extract noun/propn/verb
for token in doc:
    if token.pos_ in ["NOUN", "PROPN", "VERB"] and not token.is_stop and not token.is_punct:
        print(token.lemma_.lower())
```

```python
#task 10 location finder
for ent in doc.ents:
    if ent.label_ == 'GPE':
        print(ent.text)
```

```python
#spam keyword detector
text = 'Win a free iphone now!!! Limited offer'
doc = nlp(text)

spam_words = ["win", "free", "offer", "urgent", "limited", "prize"]

score = len(spam_words)

print("Spam score:", score)
```
