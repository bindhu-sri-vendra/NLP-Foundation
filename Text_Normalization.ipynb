```python
from num2words import num2words
```

```python

text = 'i have 25 apples and 327 oranges'
result = []
for word in text.split():
    if word.isdigit():
        result.append(num2words(int(word)).replace('-',' '))
    else:
        result.append(word)

cleaned_text = ' '.join(result)
print(cleaned_text)
```

```python
# Handling emails URL's and hashtags 
url_pattern = r'https?://\S+|www\.\S+' #matches http s is optional //. then anything or www. anything
email_pattern = r'\S+@\S+' # \s+ represents any letters followed by @ and any letters
hashtag_pattern = r'#\w+' # '#'can be followed by numbers or alphabets (\w+)
```

```python
text = "Mail me at bindhu@gmail.com and check https://example.com #AI #ML"

text = re.sub(url_pattern, '<URL>', text)
text = re.sub(email_pattern, '<EMAIL>', text)
text = re.sub(hashtag_pattern, '<HASHTAG>', text)

print(text)
```
